```python
def back(x):
    a=[]
    ins=[]
    actualt=0
    for i in x:
        ins=i.copy()
        ins[3]=ins[3]-actualt
        actualt+=ins[3]
        a.append(ins)
        
    return a
    
    
   

```

```python

def write(x,namee):
    mid=mido.MidiFile(type=1)
    track=mido.MidiTrack()
    mid.ticks_per_beat=384
    track2=mido.MidiTrack()
    track2.append(mido.MetaMessage('set_tempo',tempo=500000,time=0))
    
    track2.append(mido.MetaMessage('time_signature',numerator=4, denominator=4, clocks_per_click=24, notated_32nd_notes_per_beat=8, time=0))
    
    
    
    for i in x :
        if i[0]==1:
            
            track.append(mido.Message('note_on', note=i[1],velocity=i[2], time=i[3]))

        elif i[0]==0:
            track.append(mido.Message('note_off', note=i[1],velocity=i[2], time=i[3]))
            
    mid.tracks.append(track2)
    mid.tracks.append(track)
    mid.save(namee)
    
        
        
        
        
```
