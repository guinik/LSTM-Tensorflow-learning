```python
import mido
import numpy


def initialize(name):
    mid=mido.MidiFile(name, clip=True)
    a=[]
    fin=[]
    for i in mid.tracks[1]:

        c=str(i)
        c=c.replace('channel=0','')
        a.append(c)
    
    return a
```

```python


def second(a):
    fin=[]
    for i in a:
        if 'note_on' in i or 'note_off' in i:
            note=i.split(' ')
            res=[]
            j=0
            for b in note:
                if b=='note_on' or b=='note_off':
                    j=1

                    if b=='note_on':
                        ##LETS SUPOSE 0 OFF 1 ON FOR NOW
                        res.append(1)
                    elif b=='note_off':
                        res.append(0)


                        ##SECOND INDEX IS NOTE


                if '=' in b:
                    k=b.split('=')
                    res.append(int(k[1]))

            if j==1:
                fin.append(res)

    return fin
```

```python
def third(fin):
    finin=fin.copy()
    time=0 
    for i in finin:

        time+=i[3]
        i[3]=time

    return finin 


```
